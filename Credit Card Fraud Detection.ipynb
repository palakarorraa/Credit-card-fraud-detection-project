```python
# Splitting the dataset into features (X) and target variable (y)
```

```python
 x = data.drop(columns=['Class'])
 y = data['Class']
```

```python
# Importing the StandardScaler class from sklearn for feature scaling
```

```python
 from sklearn.preprocessing import StandardScaler
```

```python
# Standardizing the feature set (x) using StandardScaler to ensure that all features have the same scale
```

```python
 sc = StandardScaler()
 x_scaler = sc.fit_transform(x)
```

```python
 x_scaler [-1]  # Accessing the last row (sample) in the standardized feature array
```

```python
# Importing the train_test_split function from sklearn to split the dataset into training and testing sets
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the standardized data into training and testing sets
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_scaler,
    y,
    test_size=0.20,
    random_state=42, 
   stratify=y
)
```

```python
# Importing the LogisticRegression model from sklearn for binary classification
```

```python
 from sklearn.linear_model import LogisticRegression
```

```python
# Initializing the Logistic Regression model and training it on the training data
```

```python
 model = LogisticRegression()
 model.fit(x_train, y_train)
```

```python
# Making predictions on the test set using the trained Logistic Regression model
```

```python
 y_pred = model.predict(x_test)
```

```python
# Evaluating the model's accuracy on the training and testing sets
```

```python
print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))
```

```python
# Importing classification evaluation metrics from sklearn
```

```python
 from sklearn.metrics import classification_report, f1_score
```

```python
# Generating and printing the classification report to evaluate model performance on the test set
```

```python
 print(classification_report(y_test, y_pred))
```

```python
# Calculating and printing the F1 score to evaluate the model's performance on the test set
```

```python
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
# Importing the RandomForestClassifier from sklearn for classification tasks
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initializing the RandomForestClassifier model and training it on the training data
```

```python
 model1 = RandomForestClassifier()
 model1.fit(x_train, y_train)
```

```python
# Making predictions on the test set using the trained Random Forest model
```

```python
 y_pred = model1.predict(x_test)
```

```python
# Evaluating the accuracy of the trained Random Forest model on both the training and testing sets
```

```python
print("Training Accuracy :", model1.score(x_train, y_train))
print("Testing Accuracy :", model1.score(x_test, y_test))
```

```python
# Generating and printing the classification report for the Random Forest model on the test set
```

```python
 print(classification_report(y_test, y_pred))
```

```python
# Calculating and printing the F1 score for the Random Forest model on the test set
```

```python
# Importing the SMOTE (Synthetic Minority Over-sampling Technique) class from imbalanced-learn to handle class imbalance
```

```python
 from imblearn.over_sampling import SMOTE
 over_sample = SMOTE()
 x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
# Visualizing the class distribution after applying SMOTE to the training data
```

```python
y_smote.value_counts().plot(kind='bar', figsize=(15,6), color=['blue','orange'])
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# Importing the XGBClassifier from the xgboost library for classification tasks
```

```python
 from xgboost import XGBClassifier
```

```python
# Initializing the XGBoost classifier model and training it on the resampled (balanced) data using SMOTE
```

```python
 model3 = XGBClassifier(n_jobs=-1)
 model3.fit(x_smote, y_smote)
```

```python
# Making predictions on the test set using the trained XGBoost model
```

```python
 y_pred = model3.predict(x_test)
```

```python
print("Training Accuracy :", model3.score(x_train, y_train))
print("Testing Accuracy :", model3.score(x_test, y_test))
```
